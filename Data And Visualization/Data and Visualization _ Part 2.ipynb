# PREPROCESSING OF THE DATA


```python
#Copy the data from the dataframe df to another dataframe Y.
Y = data_frame.iloc[:, :].copy()

#Chages are made in the new dataframe Y.

#Query is made to have the rows which has the year from 2000 to 2016.
Y = data_frame[(data_frame['Year'] >= 2000) & (data_frame['Year'] <= 2016)]

#The columns that are not required are dropped.
Y = Y.drop(columns=['NOC', 'Games','Event'])

#Drop the rows which has the NAN values or null. 
Y = Y.dropna(how='any').reset_index()

Y
```

```python
#To check if any value is not available or null.
Y.isnull()
```

```python
##This statement will check the entire data frame and gives the output in boolean.
#The output is True even if any one of the value in the dataframe is true else false.

np.any(Y.isnull())
```

## Visualization of the data

```python
#To plot a bar graph which shows the total number of men and women who participated in olympics each year.

Y_m = Y.groupby('Year').Sex.value_counts()
Y_m
Y_m.plot(kind = 'bar' , x = 'Year', y = 'Sex')
```

```python
#To plot a stacked bar graph which shows the total number of medals won each year. 

Y_medal = Y.groupby(['Year', 'Medal'])['Year'].count().unstack('Medal')
Y_medal[['Bronze','Gold','Silver']].plot(kind='bar', stacked=True, figsize = (10,8))
```
