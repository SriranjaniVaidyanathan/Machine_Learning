## Split the data into test and train data

```python
# now partition the data 

""" partitioning data

    parameters
    -----------
    X        pd.DataFrame
             input data to partition
    T        pd.DataFrame
             target labels to partition
    raito    list
             list of ratios for partitions (should be summed to 1) 
             the number of return pairs are different
"""
def partition(X, T, ratio=[0.8, 0.2]): 
    
    assert(np.sum(ratio) == 1)
    
    # shuffle the data indices 
    idxs = np.random.permutation(X.index)
    
    # the number of samples 
    N = X.shape[0]
    
    Xs = []
    Ts = []
    i = 0  # first index to zero
    for k, r in enumerate(ratio):
        nrows = int(round(N * r))  # number of rows
        
        if k == len(ratio) -1:
            Xs.append(X.iloc[i:, :])
            Ts.append(T.iloc[i:, :])
        else:
            Xs.append(X.iloc[i:i+nrows, :])
            Ts.append(T.iloc[i:i+nrows, :])
        
        i += nrows
    
    return Xs, Ts
```

## Super Classs Definition

```python
# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    
class LinearModel(BaseModel):
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None

    # check if the matrix is 2-dimensional. if not, raise an exception    
    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):
        """
            train linear model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """        
        pass
    
    @abstractmethod
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """        
        pass 
```

## Least Squares

```python
# Linear Regression Class for least squares
class LinearRegress(LinearModel): 
    """ 
        LinearRegress class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        LinearModel.__init__(self)
        
    # train lease-squares model
    def train(self, X, T):
        # calls the add_ones function defined in the LinearModel class and is stored in Y array.
        Y = self.add_ones(X) 
        # the weight vector for the whole data set is calculated. 
        # Euclidean dot product is calculated and numpy.linalg.lstsq returns the least-squares solution to a linear matrix equation.
        self.w = np.linalg.lstsq(Y.T @ Y, Y.T @ T)[0]
        

    # apply the learned model to data X
    def use(self, X):
        # calls the add_ones function defined in the LinearModel class and is stored in Y array.
        Y = self.add_ones(X)
        # Euclidean dot product is done with Y and and the w vector
        # Y1 is the predicted value for the dependant variable in the train set. 
        Y1 = Y @ self.w
        # return the predicted value.
        return Y1
        
```

## Least Mean Square

```python
import collections # for checking iterable instance

# LMS class 
class LMS(LinearModel):
    """
        Lease Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, alpha):
        LinearModel.__init__(self)
        self.alpha = alpha
    
    # batch training by using train_step function
    def train(self, X, T):
        # N contains the number of rows in X
        N = X.shape[0]
        # loop statement to call the self.train_step function for each record.
        for i in range(N):
            self.train_step(X[i],T[i])
    
    # train LMS model one step 
    # here the x is 1d vector
    def train_step(self, x, t):
        # check if the matrix is of 2 dimentional else it reshapes the one dimentional.
        if len(x.shape)!=2:
            x = np.insert(x,0,1).reshape(-1,1)
        # Verifies if the weight vector is None else inserts a column of 1.
        if self.w is None:
            self.w = np.zeros((x.shape[0],1))
        # calculates the euclidean dot product of ((w(k)^âŠ¤)x(k)âˆ’t(k)) where T is the Transpose.
        y = self.w.T@ x- t
        # calculates the  w(k)-(alpha((w(k)^âŠ¤)x(k)âˆ’t(k)))*t.
        self.w = np.subtract(self.w, self.alpha * y * x)
        
    # apply the current model to data X
    def use(self, X):
        # predicts the value using euclidean dot product.
        t = self.w.T @ self.add_ones(X).T
        #returns the predicted value.
        return t.T
        
```

LEAST SQUARE:

- It calculates the weights after calculating it for the entire data set.
- It is similar to batch learning which generate the best predictor by learning on the entire training data set at once. 
- It is a better model if the data set is smaller.

LEAST MEAN SQUARE:
- It calculates the weight for each record and the weights are updated.
- Least square is similar to the online learning where data becomes available in a sequential order and is used to update our best predictor for future data at each step
- It is better when the data is huge and if the data is to be continously provided.
